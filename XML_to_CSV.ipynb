```python
import xml.etree.ElementTree as ET
import csv

tree = ET.parse("annotations.xml")
root = tree.getroot()
```

```python
# open a file for writing

person_data = open('Annotations-export.csv', 'w')

# create the csv writer object

csvwriter = csv.writer(person_data)
title = ['image','xmin', 'ymin', 'xmax', 'ymax', 'label']
csvwriter.writerow(title)
```

```python
for member in root.findall('image'):
    person = []
    name=member.attrib['id']+'.jpg'
    person.append(name)
    #print(name)
    for labels in member:
        if labels.attrib['label']=='head':
            person1=[]
            person1.append(person[0])
            xmin=labels.attrib['xtl']
            person1.append(xmin)
            ymin=labels.attrib['ytl']
            person1.append(ymin)
            xmax=labels.attrib['xbr']
            person1.append(xmax)
            ymax=labels.attrib['ybr']
            person1.append(ymax)
            #print(xmin, ',', ymin, ',', xmax, ',', ymax)
            a=[0,0]
            for att in labels:
                if ((att.attrib['name']=='has_safety_helmet') and (att.text=='yes')):
                    a[0]=1
                elif ((att.attrib['name']=='has_safety_helmet') and (att.text=='no')):
                    a[0]=0
                elif ((att.attrib['name']=='mask') and (att.text=='yes')):
                    a[1]=1
                else:
                    a[1]=0
            if a==[1, 1]:
                label = 'has safety_helmet and face_mask'
            elif a==[0, 1]:
                label = 'face_mask'
            elif a==[1, 0]:
                label = 'safety_helmet'
            else:
                label = 'no safety_helmet or face_mask'
            person1.append(label)
            print(person1)
            csvwriter.writerow(person1)
person_data.close()
```
